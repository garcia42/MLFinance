```python
from FinancialMachineLearning.machine_learning.clustering import clusterKMeansBase
import matplotlib.pyplot as plt
import seaborn as sns

# Clustered Mean decrease importance
corr0, clusters, _ = clusterKMeansBase(feature_analyzer.ortho_features.corr(), maxNumClusters = 10, n_init = 10)
sns.heatmap(corr0, cmap = 'viridis')
plt.savefig('{path_name}/Data/images/correlation_matrix.png')  # PNG format
plt.close()
```

```python
from FinancialMachineLearning.feature_importance.importance import mean_decrease_impurity, mean_decrease_accuracy, plot_feature_importance, cross_val_score, single_feature_importance, clustered_mean_decrease_accuracy, clustered_mean_decrease_importance_detailed, clustered_mean_decrease_accuracy

c_mdi = clustered_mean_decrease_importance_detailed(fit=feature_analyzer.fit_model, feature_names=feature_analyzer.ortho_features.columns, clstrs = clusters)

plot_feature_importance(
    c_mdi,
    oob_score=feature_analyzer.fit_model.oob_score_,
    oos_score=feature_analyzer.oos_score,
    save_fig=True,
    output_path=f'{path_name}/Data/images/Ortho_CMDI_feature_importance.png'
)
```

```python
from sklearn.metrics import accuracy_score
c_mda = clustered_mean_decrease_accuracy(model=feature_analyzer.fit_model, X=feature_analyzer.ortho_features, y = feature_analyzer.y, cv_gen=feature_analyzer.cv_folds, clusters=clusters, scoring=accuracy_score)
plot_feature_importance(
    c_mda,
    oob_score=feature_analyzer.fit_model.oob_score_,
    oos_score=feature_analyzer.oos_score,
    save_fig=True,
    output_path=f'{path_name}/images/Ortho_CMDA_feature_importance.png'
)
```
